```python
import pandas as pd
url = 'data1.csv'
dataframe = pd.read_csv(url)
dataframe.head(21)
#Luu voi loai '.csv'
```

```python
#Data location
dataframe.iloc[:10]
```

```python
#Lọc dữ liệu
dataframe = dataframe.set_index(dataframe['Name'])
dataframe.loc['Nguyen Van A']
```

```python
# Gán
dataframe[(dataframe['Faculty'] =='CKD')&(dataframe['Name']=='Nguyen Van A')]
```

```python
#Thay CKM - Cơ Khí Máy
dataframe['Faculty'].replace('CKM','Cơ Khí Máy').head(10)
```

```python
#Tìm giá trị min, max
print('Max Score:', dataframe['Score'].max())
print('Min Score:', dataframe['Score'].min())
print('Mean Score:', dataframe['Score'].mean()) # Trung binh diem
print('Total Score:', dataframe['Score'].sum()) # Tinh tong
print('The number of Student:', dataframe['Score'].count()) # De dem so nguoi
#Lay nguoi lon tuoi, nho tuoi nhat
```

```python
# Có bao nhiêu khoa
dataframe['Faculty'].unique()
# Đếm có bao nhiêu sự lặp lại
dataframe['Faculty'].value_counts()
```

```python
# Xóa dữ liệu của 1 cột
# Xóa trên bộ nhớ dữ liệu (không xóa trên file) (muốn có lại thì load lại file)
dataframe.drop('ID',axis=1).head(10) #axis = 1 là giá trị bắt buộc
dataframe.drop(['ID','Faculty'],axis=1).head(10) # [] tìm theo cột
# Xóa cột 
#dataframe.drop(dataframe.columns[0:2],axis=1).head(10)
# Xóa hàng
dataframe[dataframe['Faculty']!='CKM'].head(10)
dataframe[dataframe['Score']!='3'].head(10)
dataframe[dataframe.index != 4].head(10)
# Xóa những hàng trùng giá trị với nhau



```

```python
# Xóa những hàng trùng giá trị với nhau
dataframe.drop_duplicates().head(10)
```

```python
# Phương pháp lấy mẫu
import pandas as pd
import numpy as np
time_index = pd.date_range('01/01/2023', periods = 100000, freq = '30s')
dataframe = pd.DataFrame(index = time_index)
dataframe['Acess_Amount'] = np.random.randint(1, 10, 100000) # số nguyên, lấy ngẫu nhiên
# Lay theo tung tuan
dataframe.resample('W').sum()
```

```python
# Phương pháp lấy mẫu 2 tuần 1 lần
dataframe.resample('2W').mean()
dataframe.resample('M').mean()
```
