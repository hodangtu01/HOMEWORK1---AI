```python
from pandas.core.arrays.datetimes import conversion
# Moi quan he giua SAT va GPA
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

file = '1.01. Simple linear regression.csv'
data = pd.read_csv(file)
data.head(21)
x1 = np.array(data['SAT'])
y  = np.array(data['GPA'])


plt.scatter(x1, y)
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()


```

```python
data.hist()
plt.show()
```

```python
from sklearn import linear_model
# Tạo biến regression_model và gọi hàm
# fit là hàm học (của x, y)
# Lấy 100% dữ liệu 
regression_model = linear_model.LinearRegression()
regression_model.fit(x1.reshape(84,1), y.reshape(84,1))

# intercept hệ số giao
print(regression_model.intercept_)
print(regression_model.coef_)
# Hệ số tự do 

```

```python
#  Xác định công suất của xe phụ thuộc vào những yếu tố nào
import pandas as pd
url = 'mtcars.csv'
data = pd.read_csv(url)
# Xác định số hàng, số cột
data.shape

mpx = np.array(['mpg'])
disp = np.array(['disp'])
hp = np.array(['hp'])
wt = np.array(['wt'])


```

```python
# Hệ số tương quan
correlation_2 = data.corr()
print(correlation_2)
```

```python
data.hist()
plt.show()
```

```python
# Xóa những dữ liệu không cần thiết
X = data.drop(data.columns[[0,2]],axis = 1)
# Hàm vẽ
data.plot(kind = 'scatter',  x = 'wt', y ='mpg', color ='g', figsize =(9,9))
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
# Cho học
model.fit(X = pd.DataFrame(data['wt']), y = data['mpg'])
print(model.intercept_)
print(model.coef_)

model.score(pd.DataFrame(data['wt']),data['mpg'])
```

```python
# Du doan
y_pred = model.predict( pd.DataFrame(data['wt']))
residual = data['mpg'] - y_pred  #(y_pred = y ^)
residual.describe()
```

```python
data.plot(kind = 'scatter',  x = 'wt', y ='mpg', color ='r', figsize =(9,9))
# Chồng đồ thị
plt.plot(data['wt'], y_pred, color = 'b')
```

```python
# Bậc 2 - Hồi quy đa thức
model_2 = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**2]).T

# Cho học
model_2.fit(X, y = data['mpg'])
print('Intercept của mô hình bậc 2')
print(model_2.intercept_)
print('coeffient của mô hình bậc 2:')
print(model_2.coef_)
print('Độ chính xác của mô hình:')
print(model_2.score(X, data['mpg']))
```

```python
range = np.arange(1.5, 5.5, 0.1)
predictor_2 = pd.DataFrame([range, range**2]).T
y_pred_2 = model_2.predict(X = predictor_2)
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'b', xlim = (0, 7))
plt.plot(range, y_pred_2, color = 'y')
```

```python
model_3 = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**2, data['wt']**3]).T
# Cho học
model_3.fit(X, y = data['mpg'])
print('Intercept của mô hình bậc 3:')
print(model_3.intercept_)
print('coeffient của mô hình bậc 3:')
print(model_3.coef_)
print('Độ chính xác của mô hình:')
print(model_3.score(X, data['mpg']))
```
