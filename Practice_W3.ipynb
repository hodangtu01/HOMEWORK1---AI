```python
# Phương pháp lấy mẫu
import pandas as pd
import numpy as np
time_index = pd.date_range('01/01/2023', periods = 100000, freq = '30s')
dataframe = pd.DataFrame(index = time_index)
dataframe['Acess_Amount'] = np.random.randint(1, 10, 100000) # số nguyên, lấy ngẫu nhiên
# Lay theo tung tuan
dataframe.resample('W').sum()
```

```python
# Phương pháp lấy mẫu 2 tuần 1 lần
dataframe.resample('2W').mean()
dataframe.resample('M').mean()
```
