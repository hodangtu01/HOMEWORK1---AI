```python
# Lấy dữ liệu
from keras.datasets import cifar100
(x_train,y_train),(x_test,y_test) = cifar100.load_data()
```

```python
from keras import models
from keras.models import Sequential
from keras.layers import Dense
import numpy as np

model =  models.Sequential()
model.add(Dense(512,input_shape=(1024*3,),activation='relu'))#1024*3, 1 chiều duy nhất
model.add(Dense(100, activation = 'softmax'))
model.summary
# Tach anh mang hai chieu ve thanh 1 hang
x_train = x_train.reshape(50000,32*32*3) #28*28 = 784
x_train = x_train.astype('float32')/255
# Chuyen sang so thuc chia 255 = float32, ban dau la int

x_test = x_test.reshape(10000,32*32*3)
x_test = x_test.astype('float32')/255
```

```python
from keras.utils import to_categorical
y_train = to_categorical(y_train,100) # xem thu vien bao nhieu class
y_test  = to_categorical(y_test, 100)
```

```python
# Bien dich (compile),Train
model.compile(optimizer = 'rmsprop', loss = 'categorical_crossentropy', metrics = ['accuracy'])
model.fit(x_train, y_train, epochs = 20, batch_size = 128) 

test_loss, test_acc = model.evaluate(x_test, y_test)
print('test_accuracy',test_acc)
print('test_loss:',test_loss)

```
